```python
import pandas as pd
```

```python
df = pd.read_csv("../data/yellow_tripdata_2021-01.csv",low_memory=False)
df.head()
```

```python
df.loc[:,df.columns[df.columns.str.endswith('ID')]] = df.loc[:,df.columns.str.endswith('ID')].fillna('Unknown')

df
```

```python
## Check Duplicates
df.duplicated().sum()
```

```python
## Fill Null Values
df['VendorID'] = df['VendorID'].fillna('Unknown')
df['RatecodeID'] = df['RatecodeID'].fillna('Unknown')


df['passenger_count'] = df['passenger_count'].fillna(df['passenger_count'].median())


df['payment_type'] = df['payment_type'].fillna(df['payment_type'].mode()[0])
df['store_and_fwd_flag'] = df['store_and_fwd_flag'].fillna(df['store_and_fwd_flag'].mode()[0])

df

# Notes:
# - Mode is good for categorical values
# - Mean is good for normally distributed numerical data (no outliers)
# - Median is good for numerical data with outliers
```

```python
## Remove Outliers

def detect_outliers(df, threshold=3):
    mean_vals = df.mean()  
    std_vals = df.std()    
    
    z_scores = (df - mean_vals) / std_vals  
    return abs(z_scores) > threshold  

numeric_df = df.loc[:,['passenger_count','trip_distance','fare_amount','extra','mta_tax','tip_amount','total_amount','congestion_surcharge']]

outliers_mask = detect_outliers(numeric_df, threshold=3)

df_clean = df[~outliers_mask.any(axis=1)]

df_clean.describe()
```

```python
## # Filter Valid Transactions
df_clean = df_clean[(df_clean['fare_amount'] >= 0) & (df_clean['extra'] >= 0) & (df_clean['mta_tax'] >= 0) & (df_clean['tip_amount'] >= 0) & (df_clean['tolls_amount'] >= 0) & (df_clean['improvement_surcharge'] >= 0) & (df_clean['total_amount'] >= 0) & (df_clean['congestion_surcharge'] >= 0) ]
df_clean.head()
```
